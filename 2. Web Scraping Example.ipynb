```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
HSI_Constituents = pd.read_excel("HSIndex_Compo_Stock.xlsx")
HSI_C_name = HSI_Constituents.iloc[:, 0]
HSI_C_clean_name = []
for i in HSI_C_name:
    i = i.split(".")[0]
    # print(i)
    HSI_C_clean_name.append(i)
print(HSI_C_clean_name)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
Stock_price = []
Change = []
Percentage_change = []
Range = []
Volumn =[]
Market_Cap =[]
Turnover =[]
EPS =[]
PE_Ratio =[]
Yields =[]
Lot_Size=[]
Fivetytwo_Week = []

dict = {"Stock_code": HSI_C_clean_name[:], "Stock_price": Stock_price, 
        "Num_Change":Change,"Percentage_Change": Percentage_change,"Range":Range,'Volume':Volumn,'Market_Cap':Market_Cap
       ,"EPS":EPS, "Turnover":Turnover, "PE_Ratio":PE_Ratio,"Yield":Yields,"Lot_Size":Lot_Size,"52_week":Fivetytwo_Week}


def stock_price():
    stock_price = driver.find_element(By.CSS_SELECTOR,".abs.txt_c.ss3.cls.font-num.font-b");
    # print(stock_price.text[0:])
    stock_price = stock_price.text.split(",")[-1].replace("▲","").strip()
#    print("xxx" + stock_price + "xxx")
#    stock_price = stock_price[0].split(",")[1].replace("'","").replace("]")
    
    Stock_price.append(stock_price)

def change():
    changes = driver.find_elements(By.CSS_SELECTOR, "td.chgBox");
    changes = changes[0].text.split()[2:]
    try:
        changes = changes[0].replace("[","").replace("]","")
    except:
        pass
    #print(changes)
    Change.append(changes)
    
def percentage_change():
    percent_change = driver.find_elements(By.CSS_SELECTOR, "td.chgBox");
    percent_change = percent_change[1].text.split()[2:]
    try:
        percent_change = percent_change[0].replace("[","").replace("]","")
    except:
        pass
    #print(percent_change)
    Percentage_change.append(percent_change)
    
def ranges():
    ranges = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    ranges = ranges[0].text
    Range.append(ranges)

def volumn():
    Volumns = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    Volumns = Volumns[1].text
    Volumn.append(Volumns)

def market_cap():
    caps = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    caps = caps[2].text
    Market_Cap.append(caps)

def turnover():
    turnovers = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    turnovers = turnovers[3].text
    Turnover.append(turnovers) 

def eps():
    epss = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    epss = epss[4].text
    EPS.append(epss)   
    
def pe_ratio():
    ratios = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    ratios = ratios[5].text
    PE_Ratio.append(ratios)     
    
def yields():
    Yieldss = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    Yieldss = Yieldss[6].text
    Yields.append(Yieldss)    

def lot_size():
    sizes = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    sizes = sizes[7].text
    Lot_Size.append(sizes)   
    
def fivetytwo_week():
    weeks = driver.find_elements(By.CSS_SELECTOR, "div.ss2");
    weeks = weeks[8].text
    Fivetytwo_Week.append(weeks) 
    
for i in HSI_C_clean_name[:]:
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get("http://www.aastocks.com/en/default.aspx")
    search_input = driver.find_element("id", "sb-txtSymbol-aa")
    search_input.send_keys(i)
    search_input.send_keys(Keys.ENTER)
    
    stock_price()
    change()
    percentage_change()
    ranges()
    volumn()
    market_cap()
    turnover()
    eps()
    pe_ratio()
    yields()
    lot_size()
    fivetytwo_week()
    
    driver.close()

df = pd.DataFrame(dict)

df['Num_Change'] = df['Num_Change'].str.replace('\[\]','0')
df['Num_Change'].fillna("0", inplace=True)

df['Percentage_Change'] = df['Percentage_Change'].str.replace('\[\]','0%')
df['Percentage_Change'].fillna("0%", inplace=True)

df.loc[df['Stock_price'].str.contains("▼\n"), 'Num_Change'] = "-" + df['Num_Change']
df.loc[df['Stock_price'].str.contains("▼\n"), 'Percentage_Change'] = "-" + df['Percentage_Change']
df.loc[df['Stock_price'].str.contains("▼\n"), 'Stock_price'] = df["Stock_price"].str.replace("▼\n","")

df
```

```python
df.to_excel("HSstocks_data.xlsx",index=False)
```
